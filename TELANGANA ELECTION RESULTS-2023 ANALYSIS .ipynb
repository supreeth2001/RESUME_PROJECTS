# Party Level Analysis

# Which Parties have been present in most constituencies and Districts?

```python
vote_prty=election_data[election_data['Party']!='NOTA']
prty_cnt=vote_prty.groupby('Party').apply(lambda x:x['CONSTITUENCY'].count()).reset_index(name='# Constituency')
prty_st=vote_prty.groupby('Party').apply(lambda x:x['District'].nunique()).reset_index(name='# District')
prty_cnt.sort_values(by='# Constituency',ascending=False,inplace=True)
prty_top_cn=prty_cnt[:25]
prty_top_all=pd.merge(prty_top_cn,prty_st,how='inner',left_on='Party',right_on='Party')
fig = px.scatter(prty_top_all, x='# Constituency', y='# District', color='# District',
                 size='# Constituency', hover_data=['Party'])
fig.update_layout(title_text='Constituency vs Districtwise participation for the most contesting Political Parties',template='plotly_dark')
fig.show()
```

 **Observation**  Indian National Congress (INC) and Bharat rashtra samiti have participated in the most number of constituencies all over telangana. While INC leads in the number of constituency contested, INC wins in terms of the number of Districts. While these are the major parties to contest almost all over Telangana, we see the rest of the parties have restricted themselves to a handfull of states.

# What has been the performance of the Parties Districtwise?

```python
dt_prty=vote_prty.groupby(['Party','District']).apply(lambda x:x['WIN'].sum()).reset_index(name='WIN')
pvt_st_prty=pd.pivot(dt_prty,index='Party',columns='District',values='WIN')
plt.figure(figsize=(15,15))
sns.heatmap(pvt_st_prty,annot=True,fmt='g',cmap='terrain')
plt.xlabel('District',size=30)
plt.ylabel('Party',size=30)
plt.title('Districtwise report card for the Political Parties in Telangana',size=30)
```

# Which party has won the most constituencies?

```python
part_win=election_data.groupby('Party').apply(lambda x:x['WIN'].sum()).reset_index(name='# WIN')
part_win.sort_values(by='# WIN',ascending=False,inplace=True)
top_part_win=part_win[0:15]
fig = px.bar(top_part_win, x='Party', y='# WIN',
                     color='# WIN',title='Win Counts by a Political Party in 2023')
fig.update_layout(title_text='Win Counts by a Political Party in 2023',template='plotly_dark')
fig.show()
```

# What has been the general Win vs Loss relationship for the Parties in 2023?

```python
prty_cnt_win=pd.merge(prty_cnt,part_win,how='inner',left_on='Party',right_on='Party')
prty_cnt_win['Lost']=prty_cnt_win['# Constituency']-prty_cnt_win['# WIN']
prty_wins_cnt=prty_cnt_win[['Party','# WIN']]
prty_wins_cnt['Verdict']='Constituency Won'
prty_loss_cnt=prty_cnt_win[['Party','Lost']]
prty_loss_cnt['Verdict']='Constituency Lost'
prty_wins_cnt.columns=['Party','Counts','Verdict']
prty_loss_cnt.columns=['Party','Counts','Verdict']
top_prty_wins_cnt=prty_wins_cnt[:15]
prty_loss_cnt_cnt=prty_loss_cnt[:15]
prt_win_loss=pd.concat([top_prty_wins_cnt,prty_loss_cnt_cnt])
fig = px.bar(prt_win_loss, x='Party', y='Counts', color='Verdict')
fig.update_layout(title_text='Win vs Loss Analysis for the Top Parties',template='plotly_dark')
fig.show()
```

# Politician Level Analytics

# What relation does the Politician category have with the election results?

```python
vote_gndr=election_data[election_data['Party']!='NOTA']
cat_overall=vote_gndr.groupby('GENDER').apply(lambda x:x['Winning Candidate'].count()).reset_index(name='Counts')
cat_overall['Category']='Overall GENDER Counts'
winners_cat=vote_gndr[vote_gndr['WIN']==1]
cat_winner=winners_cat.groupby('GENDER').apply(lambda x:x['Winning Candidate'].count()).reset_index(name='Counts')
cat_winner['Category']='Winning GENDER Ratio'
cat_overl_win=pd.concat([cat_winner,cat_overall])
fig = px.bar(cat_overl_win, x='GENDER', y='Counts',
             color='Category', barmode='group')
fig.update_layout(title_text=' Win Counts for the Category in Politics',template='plotly_dark')
fig.show()
```

**Observation**.out of the total list of participants only **19.83%(6 out of 119)** are female,while **80.17%(113 out of 119)** are male . 

 Upon considering the winners, 19.83%(6 out of 119) are female politicians, while 85.9% are male politicians. 
 
 The Gender ratio is not very well distributed as can be seen from the above presentation.
